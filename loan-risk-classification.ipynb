```python
data_encoded = pd.get_dummies(data, columns=['person_home_ownership'], drop_first=True)
print(data_encoded.head())
```

```python
data_encoded.person_gender = [1 if value == "male" else 0 for value in data.person_gender]
data_encoded.drop(['person_education','loan_intent'], axis=1, inplace = True)
data_encoded.head()
```

```python
data_encoded.previous_loan_defaults_on_file	 = [1 if value == "No" else 0 for value in data_encoded.previous_loan_defaults_on_file	]
```

```python
y = data_encoded.loan_status
x = data_encoded.drop(['loan_status'], axis=1)
```

```python
y
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaler = scaler.fit_transform(x)
```

```python
x_scaler
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaler, y, test_size=0.3, random_state=40)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import classification_report
cl = classification_report(y_test, y_pred)
print(cl)
```
